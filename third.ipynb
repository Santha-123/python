```python
#arbitary arguments
#*args example
def order_food(*items):
    cart= []
    for item in items:
        cart.append(item)
    return cart
```

```python
#arbitary arguments
#*args example for calling the above function
lunch=order_food('chilly chicken','garlic chicken','chicken lollipop')
print(lunch)
```

```python
#arbitary arguments
#*args example
def sum_all_numbers(*numbers):
    total = sum(numbers)
    return total
```
